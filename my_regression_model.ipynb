```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
housing_ds = pd.read_csv("housing.csv")
```

```python
#getting familiar the data
housing_ds.head()
```

```python
housing_ds.info()
```

```python
housing_ds.describe()
```

```python
X = housing_ds.drop("median_house_value", axis=1)
y = housing_ds["median_house_value"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

```python
#TRAINING MY MODEL
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
X_test.head()
```

```python
X_test.drop("ocean_proximity", axis=1, inplace=True)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")
imputer.fit(X_test)

transformed_test = imputer.fit_transform(X_test)
```

```python
#predict
predd = lin_reg.predict(transformed_test)
```

```python
predd
```

```python
y_test
```

```python
#model performance
from sklearn.metrics import mean_squared_error

lin_mse = mean_squared_error(y_test, predd)
lin_rmse = np.sqrt(lin_mse)

```
